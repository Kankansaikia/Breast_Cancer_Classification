```python
from sklearn.datasets import load_breast_cancer
breast_cancer=load_breast_cancer()
```

```python
print("Target names of the dataset: ",breast_cancer['target_names'])
```

```python
print("Feature names of the dataset: ",breast_cancer['feature_names'])
```

```python
print("Shape of the dataset: ",breast_cancer['target'].shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(breast_cancer['data'],breast_cancer['target'],random_state=1)
```

```python
print("Shape of training data: ",X_train.shape)
print("Shape of outcome of training data: ",Y_train.shape)
```

```python
print("Shape of testing data: ",X_test.shape)
print("Shape of outcome of testing data: ",Y_test.shape)
```

```python
import pandas as pd
dataframe=pd.DataFrame(X_train,columns=breast_cancer.feature_names)
```

```python
dataframe.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  iris_dataframe is your DataFrame and Y_train is your target array
df = dataframe.copy()
df['target'] = Y_train

# Plot histograms for each feature
features = dataframe.columns
num_features = len(features)
rows = (num_features // 2) + (num_features % 2)
colors = ['b', 'g']  # There are 3 classes in Y_train

plt.figure(figsize=(20, 30))
for i, feature in enumerate(features):
    plt.subplot(rows, 2, i + 1)  # Adjust subplot grid based on the number of features
    for target_value in df['target'].unique():
        subset = df[df['target'] == target_value]
        plt.hist(subset[feature], bins=20, alpha=0.6, label=f'Class {target_value}', color=colors[target_value])
    plt.title(feature)
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.legend()

plt.suptitle('Histograms of Each Feature Grouped by Target', y=1.02)
plt.tight_layout()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
train_accuracy=[]
test_accuracy=[]
iterator=range(1,11)
for i in iterator:
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    train_acc=knn.score(X_train,Y_train)
    train_accuracy.append(train_acc)
    test_acc=knn.score(X_test,Y_test)
    test_accuracy.append(test_acc)
train_accuracies = ["{:.4f}".format(acc) for acc in train_accuracy]
test_accuracies = ["{:.4f}".format(acc) for acc in test_accuracy]

print("The training accuracies are: ", train_accuracies)
print("The test accuracies are: ", test_accuracies)
    
```

```python
import matplotlib.pyplot as plt
plt.plot(iterator,train_accuracy,label="Training_Accuracy")
plt.plot(iterator,test_accuracy,label="Test_Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Iterator")
plt.legend()
```
